```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.randint(1, 100, size=100)
```

## Graph

```python
plt.plot(nums)
plt.title("Graph of Random Numbers")
plt.legend("nums")
plt.show()
```

## Subplot

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

## Horizontal Bar Chart

```python
plt.barh(nums, nums)
plt.title("Bar Chart for Random Values")
plt.show()
```
