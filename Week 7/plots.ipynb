```python
import matplotlib.pyplot as plt
```

```python
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
populations = [99.2, 100.6, 102.1, 102.8, 103.4, 104.0, 105.1, 106.5, 107.0, 108.2]
```

## Scatter Plot

```python
plt.scatter(years, populations)
plt.show()
```

## Bar Plot

```python
plt.bar(years, populations)
plt.show()
```

## Line Graph

```python
plt.plot(years, populations)
plt.show()
```
