### Q3: Show the apps having rating above than 4.0 and have maximum downloads.

```python
psdata1['Installs'] = psdata1.Installs.str.replace(",","")
psdata1['Installs'] = psdata1.Installs.str.replace("+", "")
psdata1['Installs'] = psdata1.Installs.str.replace("Free", np.nan)
```

```python
psdata1['Installs'] = psdata1.Installs.astype(float)
```

```python
psdata1[(psdata1['Rating']>=4.0)&(psdata1['Installs']==psdata1.Installs.max())]
```
