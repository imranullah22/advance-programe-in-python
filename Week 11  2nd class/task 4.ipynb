```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('population2020.csv')
data
```

```python
data.set_index('Country', inplace=True)
```

```python
data
```

```python
data.isnull()
```

```python
data = data.dropna()
```

```python
data
```

```python
top3 = data.nlargest(3, ['Population'])
```

```python
top3
```

```python
plt.pie(top3['Population'].values, labels=top3['Country'].values)
plt.show()
```

```python
plt.pie(data.nsmallest(3, ['Population'])['Population'].values, labels=data.nsmallest(3, ['Population'])['Country'].values)
plt.show()
```
